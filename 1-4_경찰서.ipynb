```python
seoul = pd.read_excel("../Report.xls")
seoul = seoul.iloc[2:,]
seoul = seoul.rename(columns={'자치구':'지역'})
seoul
```

```python
seoul["면적"] = pd.to_numeric(seoul["면적"])
```

```python
gu_police
```

```python
gu_police = gu_police.rename(columns={'구':'지역'})
```

```python
seoul_police = pd.merge(gu_police, seoul, on='지역')
seoul_police = seoul_police[['지역','경찰서 갯수','면적']]
seoul_police
```

```python
seoul_police["면적비율"] = seoul_police["경찰서 갯수"] / seoul_police["면적"] * 100
seoul_police
```

```python
seoul_police.rename(columns={'지역':'구', '경찰서 갯수':'경찰서갯수', '면적':'구면적'}, inplace=True)
gu_police = seoul_police
gu_police["면적비율"] = round(seoul_police["면적비율"], 2)
gu_police.to_csv("../df_gu_police.csv", index=None)
gu_police
```

```python
# 구별 공원면적

police = gu_police.sort_values(by=['면적비율'], ascending=True)
ax1 = police.plot.barh(figsize=(12,10), x='구', y='면적비율', xlabel='', ylabel='면적비율', 
                     color=plt.cm.tab20(np.arange(10)), title='구별 경찰서 면적비율'
                     )

# annotate(표시할값, x좌표, y좌표, 옵션)
for p in ax1.patches: 
    ax1.annotate(p.get_width(), (p.get_x()+p.get_width(), p.get_y()+0.05) )
ax1
```
