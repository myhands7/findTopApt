## 01 공원

```python
# 구별 공원면적
gu_park = pd.read_csv("df_gu_park.csv")

gu_park = gu_park.sort_values(by=['공원면적'], ascending=True)
ax = gu_park.plot.barh(figsize=(9,7), x='구', y='공원면적', xlabel='', ylabel='공원면적', 
                       color=plt.cm.tab20(np.arange(10)), title='구별 공원면적'
                       )

# annotate(표시할값, x좌표, y좌표, 옵션)
for p in ax.patches: 
    ax.annotate(p.get_width(), (p.get_x()+p.get_width(), p.get_y()+0.05) )
ax
```

```python
# 구별 면적대비 공원면적
gu_park = pd.read_csv("df_gu_park.csv")


gu_park = gu_park.sort_values(by=['면적비율'], ascending=True)
ax1 = gu_park.plot.barh(figsize=(9,7), x='구', y='면적비율', xlabel='', ylabel='면적비율', 
                        color=plt.cm.tab20(np.arange(10)), title='구별 면적대비 공원면적'
                        )

# annotate(표시할값, x좌표, y좌표, 옵션)
for p in ax1.patches: 
    ax1.annotate(p.get_width(), (p.get_x()+p.get_width(), p.get_y()+0.05) )
ax1
```
