```python
gu_hospital = gu_hospital.rename(columns={'구':'지역'})
gu_hospital
```

```python
seoul = pd.read_excel("../Report.xls")
seoul = seoul.iloc[2:,]
seoul = seoul.rename(columns={'자치구':'지역'})
seoul
```

```python
seoul["면적"] = pd.to_numeric(seoul["면적"])
```

```python
seoul_hospital = pd.merge(gu_hospital, seoul, on='지역')
seoul_hospital = seoul_hospital[['지역', '병원 갯수', '면적']]
seoul_hospital
```

```python
seoul_hospital["면적비율"] = seoul_hospital["병원 갯수"] / seoul_hospital["면적"] * 100
seoul_hospital
```

```python
seoul_hospital.rename(columns={'지역':'구', '병원 갯수':'병원갯수', '면적':'구면적'}, inplace=True)
gu_hospital = seoul_hospital
gu_hospital["면적비율"] = round(seoul_hospital["면적비율"], 2)
gu_hospital.to_csv("../df_gu_hospital.csv", index=None)
gu_hospital
```

```python
# 구별 공원면적

hospital = gu_hospital.sort_values(by=['구면적'], ascending=True)
ax1 = hospital.plot.barh(figsize=(12,10), x='구', y='면적비율', xlabel='', ylabel='면적비율', 
                     color=plt.cm.tab20(np.arange(10)), title='구별 병원수 면적비율'
                     )

# annotate(표시할값, x좌표, y좌표, 옵션)
for p in ax1.patches: 
    ax1.annotate(p.get_width(), (p.get_x()+p.get_width(), p.get_y()+0.05) )
ax1
```
